```python
X, y = generate_data()
plt.scatter(X[:, 0], X[:, 1], s=40, c=y, cmap=plt.cm.Spectral)
plt.show()
```

## e) Train network with different activation functions
### Tanh


```python
model = NeuralNetwork(nn_input_dim=2, nn_hidden_dim=3 , nn_output_dim=2, actFun_type='tanh')
model.fit_model(X,y,epsilon=0.01)
model.visualize_decision_boundary(X,y)
```

### Sigmoid

```python
model = NeuralNetwork(nn_input_dim=2, nn_hidden_dim=3 , nn_output_dim=2, actFun_type='sigmoid')
model.fit_model(X,y,epsilon=0.01)
model.visualize_decision_boundary(X,y)
```

### ReLU

```python
model = NeuralNetwork(nn_input_dim=2, nn_hidden_dim=3 , nn_output_dim=2, actFun_type='relu')
model.fit_model(X,y)
model.visualize_decision_boundary(X,y)
```

### More hidden units with Tanh

```python
model = NeuralNetwork(nn_input_dim=2, nn_hidden_dim=20 , nn_output_dim=2, actFun_type='tanh')
model.fit_model(X,y)
model.visualize_decision_boundary(X,y)
```
